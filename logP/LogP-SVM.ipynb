```python
import pandas as pd

from tqdm import tqdm

from sklearn import svm
from sklearn import neighbors
from sklearn import preprocessing
from sklearn import model_selection, linear_model
from sklearn import feature_selection
```

```python
df = pd.read_csv('logP.csv')
df
```

```python
y = df.logP
X = df.iloc[:, 3:]
```

```python
kf = model_selection.RepeatedKFold(n_repeats=5, n_splits=5, random_state=42)
```

```python
X_norm = preprocessing.StandardScaler().fit_transform(X)
```

```python
X_new = pd.DataFrame(X_norm, columns = X.columns)
X_new
```

```python
X_norm.shape
```

## SVM

```python
select_svm = feature_selection.SequentialFeatureSelector(svm.SVR(), n_jobs=-1, direction='backward', cv=kf)
select_svm.fit(X_new, y)
```

```python
select_svm.n_features_to_select_
```

```python
X_sel_feat_svm_names = X_new.columns[select_svm.get_support()]
X_sel_feat_svm = X_new[X_sel_feat_svm_names]
X_sel_feat_svm
```

```python

model_selection.cross_val_score(linear_model.LinearRegression(), X_sel_feat_svm, y, cv=kf).mean()
```

## kNN (KNN)

```python
def knn_counter(X_new, y):
    
    results = {}
 
    for i in tqdm(range(1, 150, 20)):

        select_kNN = feature_selection.SequentialFeatureSelector(neighbors.KNeighborsRegressor(n_neighbors = i), n_jobs=-1, cv=kf)
        select_kNN.fit(X_new, y)

        X_sel_feat_knn_names = X_new.columns[select_kNN.get_support()]
        X_sel_feat_knn = X_new[X_sel_feat_knn_names]

        results[i] = model_selection.cross_val_score(linear_model.LinearRegression(), X_sel_feat_knn, y, cv=kf)

    return sorted(results.item(), key = lambda x: x[1])[0]
```

```python
knn_score1 = knn_counter(X_new, y)
```

```python
select_kNN = feature_selection.SequentialFeatureSelector(neighbors.KNeighborsRegressor(n_neighbors = 2), n_jobs=-1, cv=kf)
select_kNN.fit(X, y)
```

```python
select_kNN.n_features_to_select_
```

```python
X_sel_feat_knn_names = X_new.columns[select_kNN.get_support()]
X_sel_feat_knn = X_new[X_sel_feat_knn_names]
X_sel_feat_knn
```

```python

model_selection.cross_val_score(linear_model.LinearRegression(), X_sel_feat_knn, y, cv=kf)
```
