```python
smiles = [
    "CCCC",  # CH3-CH2-CH2-CH3
    "CF",  # CH3-F
    "FC",  # F-CH3
    "CC(=O)O",  # CH3-COOH
    "CC(O)=O",  # ...
    "OC(=O)C",  # ...
    "[Na+].[Na+].[O-]S([O-])(=O)=O",  # Na2SO4
]
```

```python
dm.to_image(smiles, legends=smiles)
```

```python
for smi in smiles:
    mol = dm.to_mol(smi)
    canon_smi = dm.to_smiles(mol, canonical=True)
    print(smi, canon_smi)
```
